```python
#Importing the necessary libaries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Function to draw a bar graph.
#Note: I have returned an object for some flexibility.
def get_bargraph(x_values,y_values,graph_title,x_label,y_label):
    fig, ax = plt.subplots()
    width = 0.75  # the width of the bars
    ind = np.arange(len(y_values))  # the x locations for the groups
    ax.barh(ind, y_values, width, color="blue")
    ax.set_yticks(ind + width / 2)
    ax.set_yticklabels(x_values, minor=False)
    for i, v in enumerate(y_values):
        ax.text(v, i, " " + str(v), color='blue', va='center', fontweight='bold')
    plt.title(graph_title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plot = plt.show()
    return plot
```

```python
#Function to draw pie chart.
#Note: I have returned an object for some flexibility.
def get_piechart(values,labels):
    fig = plt.figure(figsize=(10, 7))
    p, tx, autotexts = plt.pie(values, labels=labels, autopct="", shadow=True)
    for i, a in enumerate(autotexts):
        a.set_text("{}".format(values[i]))
    plt.axis('equal')
    plot = plt.show()
    return plot
```

```python
#look at the data and understand how it is structured.
df = pd.read_csv(r'12.csv')
print(df)
```

```python
#Create empty dictionary to save the data
dic ={}
i= 0
#Loop through the data and save the squad name as a key and the data associated with that team as a list of values.
for team in df["Squad"]:
    dic[team] = []
    #I have got only data from columns 3 through 12(index starts at 0).
    for column in df.columns[3:13]:
        dic[team].append(df[column][i])
    i = i+1
#Save the number of goals each team has scored in a list to pass it as an argument.
z = [dic[x][3] for x in dic.keys()]
print(dic)
print()
print(z)
```

```python
#Now we call the function and pass the required arguments.
get_bargraph(dic.keys(),z,"Goals per team season 2020-2021","Goals","Teams")
```

```python
#Let's get how many times each team has won and save it to a list.
won = [dic[x][0] for x in dic.keys()]
print(won)
```

```python
#Let's visualize that using pie chart.
get_piechart(won,dic.keys())
```
