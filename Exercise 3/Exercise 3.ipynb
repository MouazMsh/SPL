```python
#Importing the neccacary libaries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Function to display values on the graph.
def display_values(x_values,y_values):
    for x, y in zip(x_values, y_values):
        label = "{:}".format(y)

        plt.annotate(label,  # this is the text
                     (x, y),  # this is the point to label
                     textcoords="offset points",  # how to position the text
                     xytext=(0, 10),  # distance from text to points (x,y)
                     ha='center')  # horizontal alignment can be left, right or center
```

```python
#Function to draw line graph.
def get_lineGraph(x_values,y_values,x_label,y_label,graph_title,team_name):
    df = pd.DataFrame({'x_values': x_values, 'y1_values':y_values})
    plt.plot('x_values', 'y1_values', data=df, marker='o', color='skyblue', linewidth=2, label=team_name)
    plt.legend(loc='lower left')
    plt.title(graph_title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.yticks(np.arange(350, 500, step=25))
    plt.grid(True)
    display_values(x_values, y_values)
    plt.rcParams["figure.figsize"] = (9,5)
    plt.show()
```

```python
#Let's get the data, and save them in a dictionary.
shooting_dic ={}
df = pd.read_csv(r'Shooting-2018-2019.csv')
i=1
for team in df["Unnamed: 0"][1:]:
    shooting_dic[team] = {}
    shooting_dic[team]["2018-2019"] = []
    for column in df.columns[3:11]:
        shooting_dic[team]["2018-2019"].append(df[column][i])
    i +=1


df = pd.read_csv(r'Shooting-2019-2020.csv')
i = 1
for key in shooting_dic.keys():
    shooting_dic[key]["2019-2020"] = []
    for column in df.columns[3:11]:
        shooting_dic[key]["2019-2020"].append(df[column][i])
    i +=1

df = pd.read_csv(r'Shooting-2020-2021.csv')
i = 1
for key in shooting_dic.keys():
    shooting_dic[key]["2020-2021"] = []
    for column in df.columns[3:11]:
        shooting_dic[key]["2020-2021"].append(df[column][i])
    i +=1
print(shooting_dic)
```

```python
#Let's get what we want for now and that it is the number of shoots Al-Ahli has done for the last three seasons.
seasons = ["2018-2019","2019-2020","2020-2021"]
shoots_ahli = [int(shooting_dic[key][season][1]) for key in list(shooting_dic.keys())[:1] for season in shooting_dic[key]]
```

```python
#Let's visualize that.
get_lineGraph(seasons,shoots_ahli,"Seasons","Shoots","Shoots Vs Seasons","Al-Ahli")
```

```python
#Let's do it for the rest big four teams.
shoots_hilal = [int(shooting_dic[key][season][1]) for key in list(shooting_dic.keys())[1:2] for season in shooting_dic[key]]
get_lineGraph(seasons,shoots_hilal,"Seasons","Shoots","Shoots Vs Seasons","Al-Hilal")
```

```python
shoots_ittihad = [int(shooting_dic[key][season][1]) for key in list(shooting_dic.keys())[2:3] for season in shooting_dic[key]]
get_lineGraph(seasons,shoots_ittihad,"Seasons","Shoots","Shoots Vs Seasons","Al-Ittihad")
```

```python
shoots_nasr = [int(shooting_dic[key][season][1]) for key in list(shooting_dic.keys())[3:4] for season in shooting_dic[key]]
get_lineGraph(seasons,shoots_nasr,"Seasons","Shoots","Shoots Vs Seasons","Al-Nasr")
```

```python
shoots_shabab = [int(shooting_dic[key][season][1]) for key in list(shooting_dic.keys())[4:] for season in shooting_dic[key]]
get_lineGraph(seasons,shoots_shabab,"Seasons","Shoots","Shoots Vs Seasons","Al-Shabab")
```

```python
#Let's get the shots on target's accuracy for the big five teams.
sp_ahli = [float(shooting_dic[key][season][3]) for key in list(shooting_dic.keys())[:1] for season in shooting_dic[key]]
sp_hilal = [float(shooting_dic[key][season][3]) for key in list(shooting_dic.keys())[1:2] for season in shooting_dic[key]]
sp_ittihad = [float(shooting_dic[key][season][3]) for key in list(shooting_dic.keys())[2:3] for season in shooting_dic[key]]
sp_nasr = [float(shooting_dic[key][season][3]) for key in list(shooting_dic.keys())[3:4] for season in shooting_dic[key]]
sp_shabab = [float(shooting_dic[key][season][3]) for key in list(shooting_dic.keys())[4:] for season in shooting_dic[key]]
```

```python
#Let's visualize that using line chart.
df=pd.DataFrame({'x_values': seasons, 'y1_values':sp_ahli, 'y2_values': sp_hilal, 'y3_values': sp_ittihad, 'y4_values': sp_nasr, 'y5_values': sp_shabab})
plt.plot( 'x_values', 'y1_values', data=df, marker='o', color='skyblue', linewidth=2, label="Ahli")
plt.plot( 'x_values', 'y2_values', data=df, marker='o', color='red', linewidth=2, label="Hilal")
plt.plot( 'x_values', 'y3_values', data=df, marker='o', color='olive', linewidth=2, label="Ittihad")
plt.plot( 'x_values', 'y4_values', data=df, marker='o', color='yellow', linewidth=2, label="Nasr")
plt.plot( 'x_values', 'y5_values', data=df, marker='o', color='orange', linewidth=2, label="Shabab")
plt.legend()
plt.title('Shot\'s on goal accuracy percent Vs Seasons')
plt.xlabel('Seasons')
plt.ylabel('Shot\'s on goal accuracy percent')
plt.grid(True)
plt.rcParams["figure.figsize"] = (20,6)
display_values(seasons,sp_ahli)
display_values(seasons,sp_hilal)
display_values(seasons,sp_ittihad)
display_values(seasons,sp_nasr)
display_values(seasons,sp_shabab)
plt.show()
```
